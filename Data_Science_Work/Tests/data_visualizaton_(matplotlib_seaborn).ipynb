```python
"""
 Que 2.	Visualize the distribution of test scores for students from three different schools.
•	Data includes score and school columns.
•	Use a box plot to compare distributions.
"""
# Solution :
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt 

data ={
   'score':[30,40,50,60,70,80,90],
   'school':['A','B','B','C','C','C','C']
}
df=pd.DataFrame(data)
sns.boxplot(x='school', y='score', data=df)
plt.title('Test S.Score ')
plt.xlabel('school')
plt.ylabel('score')
plt.show()


```

```python
'''
QUE 4.	Create a line chart showing the monthly average temperature of a city over a year.
•	x-axis: Months (January to December)
•	y-axis: Temperature (in Celsius)
•	Add a title, axis labels, and gridlines.
•	Customize the line style (e.g., dashed, color, markers)
'''
# Solution:
import pandas as pd
import matplotlib.pyplot as plt
months = ['J','F','M','A','May','Ju',
          'Jul','Au','S','O','N','D']
temperature = [4,5,9,14, 19,25,27,16,22,35, 10,12]  
plt.figure(figsize=(8, 4))
plt.plot(months,temperature,linestyle='--', color='blue', marker='o',alpha=1)
plt.title('Monthly avg temp of city')
plt.xlabel('Month')
plt.ylabel('Temperature in degree celcius')
plt.grid(True)
plt.show()
```
